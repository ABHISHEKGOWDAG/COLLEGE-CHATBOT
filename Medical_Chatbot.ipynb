```python
from langchain import PromptTemplate # import PromptTemplate

prompt_template="""
Use the following pieces of information to answer the user's question.
If you don't know the answer, just say that you don't know, don't try to make up an answer.

Context: {context}
Question: {question}

Only return the helpful answer below and nothing else.
Helpful answer:
"""

PROMPT=PromptTemplate(template=prompt_template, input_variables=["context", "question"])
chain_type_kwargs={"prompt": PROMPT}
```

```python
!pip install ctransformers
!pip install langchain-community # Install the langchain-community package

from langchain.llms import CTransformers # Import the CTransformers class from langchain.llms
from huggingface_hub import hf_hub_download
# Download and save the model locally
model_path = hf_hub_download(repo_id="TheBloke/Llama-2-7B-Chat-GGML",
                             filename="llama-2-7b-chat.ggmlv3.q4_0.bin")

# Now use the local path to the model
llm = CTransformers(
    model = model_path,
    model_type="llama",
    config={"max_new_tokens":512, "temperature":0.8}
)
```

```python
!pip install langchain --upgrade # ensure langchain is up to date
from langchain.chains import RetrievalQA # import the RetrievalQA object from langchain.chains

qa = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever= docsearch.as_retriever(search_kwargs={"k":2}),
    return_source_documents = True,
    chain_type_kwargs=chain_type_kwargs
)
```

```python
while True:
    user_input = input(f"Input Prompt: ")
    result = qa({"query":user_input})
    print("Response:", result["result"])
```
